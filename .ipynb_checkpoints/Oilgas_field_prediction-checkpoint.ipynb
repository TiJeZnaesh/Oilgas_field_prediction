## Обработка пропущенных значений

```python
#заполнение пропусков с помощью случайного леса
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import random
import numpy as np
random.seed(42)
np.random.seed(42)

def RandomForest_imputer(df):
    if len(df.columns[df.isna().any()].tolist()) == 1:
        col = df.loc[:, df.isna().sum() > 0].columns[0]

        train = df.dropna().copy()
        test = df[df[col].isna()].copy()

        col_index = df.columns.get_loc(col)

        y_train = train[col]
        X_train = train.drop(col, axis=1)

        test = test.drop(col, axis=1)

        model = RandomForestRegressor()
        model.fit(X_train, y_train)

        y_pred = model.predict(test)
        test.insert(loc=col_index, column=col, value=y_pred)

        df = pd.concat([train, test])
        df.sort_index(inplace=True)

        return df
    else:
        missing_columns = df.columns[df.isna().any()].tolist()
        # заполнить пропуски случайными значениями и сохранить в df_imp в "имя_признака + '_imp'"
        df_imp = df.copy()
        df_result = df.copy()

        for feature in missing_columns:
            df_imp[feature + '_imp'] = df_imp[feature]
            number_missing = df_imp[feature].isnull().sum()
            observed_values = df_imp.loc[df_imp[feature].notnull(), feature]
            df_imp.loc[df_imp[feature].isnull(), feature + '_imp'] = np.random.choice(observed_values, number_missing,
                                                                                      replace=True)

        for feature in missing_columns:
            # Параметры: признаки - неполные признаки - признак для предсказания + '_imp'
            parameters = list(set(df_imp.columns) - set(missing_columns) - {feature + '_imp'})

            # X - Параметры, Y - признак для предсказания + '_imp'
            model = RandomForestRegressor()
            model.fit(X=df_imp[parameters], y=df_imp[feature + '_imp'])

            df_result.loc[df_imp[feature].isnull(), feature] = model.predict(df_imp[parameters])[
                df_imp[feature].isnull()]

        return df_result
```

```python
train = RandomForest_imputer(train)
```

```python
train.isna().sum()
```

```python
test = RandomForest_imputer(test)
```

```python
test.isna().sum()
```

# Обучение модели

```python
#для обучения модели сразу делаем вектор признаков(Х)фьючерс и вектор ответов(у)таргет
X = train.drop(columns = "Onshore/Offshore")
y = train["Onshore/Offshore"]
```

```python
#разделим на подвыборки(или можно использовать кросс-валидацию) одно и тоже
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
#решаем эту задачу с помощью дерева решений
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
#обучаем теперь вот эту созданную модель и предсказываем ответы в тесте только по признакам
model.fit(X_train, y_train)
answers = model.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
#обучаем теперь вот эту созданную модель и предсказываем ответы в тесте только по признакам
model.fit(X_train, y_train)
answers = model.predict(X_test)
```

```python
#проверим по метрикам работу нашей модели
from sklearn.metrics import accuracy_score, precision_score, recall_score
print(accuracy_score(y_test, answers))
```

## Формирование ответов

```python

y_pred_test = model.predict(test)

y_pred_test = pd.DataFrame(y_pred_test, columns=['Onshore/Offshore'])
y_pred_test['Onshore/Offshore'] = y_pred_test['Onshore/Offshore'].replace(save)
y_pred_test = y_pred_test.reset_index()
y_pred_test.to_csv("solution.csv", index = False)
```
